```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error
```

```python
json_data = pd.read_json('tiktok_dataset_100000.json')
df = pd.DataFrame(json_data)
```

```python
df.head()
```

```python
df.drop(['Create_time'], axis=1, inplace=True)
df.head()
```

```python
# Split the data into features (X) and target variable (y)
y = df['Views']
X = df.drop('Views', axis=1)

X.head()
```

```python
scaler = MinMaxScaler()
X_normalized = scaler.fit_transform(X)
```

```python
# Split the data into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X_normalized, y, test_size=0.2, random_state=42)
```

```python

# Create the Random Forest Regression model
rf_model = RandomForestRegressor(n_estimators=100, random_state=42)

# Train the model
rf_model.fit(X_train, y_train)
```

```python
import joblib

# Save the model
joblib.dump(rf_model, 'tiktok_rf_100k.pkl')
```

```python
rf_model.predict(X_test)
```

```python
# Make predictions on the test set
y_pred = rf_model.predict(X_test).round()

# Calculate the mean squared error
mse = mean_squared_error(y_test, y_pred)
mse
```

```python
r_squared = rf_model.score(X_test, y_test)
r_squared
```
